```python
# Remove all extra newline characters (\n) from the 'Team & Contract' column
df['Team & Contract'] = df['Team & Contract'].str.replace(r'\n+', ' ', regex = True)
print(df['Team & Contract'])
```

```python
# Remove all leading and trailing spaces from the 'Team & Contract' column
df['Team & Contract'] = df['Team & Contract'].str.strip()

df.head()
```

```python
# Split the 'Team & Contract' column into 'Team' and 'Contract' columns
df[['Team', 'Contract']] = df['Team & Contract'].str.split(r' (?=\d{4})', n=1, expand=True)

df['Team'] = df['Team'].str.strip()
df['Contract'] = df['Contract'].str.strip()
df['Contract'] = df['Contract'].str.replace('~', '-')

df
```

```python
# Modify column names
df.rename(columns={'↓OVA': 'Overall Rating', 'Height': 'Height(ft)', 'Weight': 'Weight(lbs)', 'Wage': 'Wage(€)'}, inplace=True)
```

```python
relevant_columns = ['Overall Rating', 'Name', 'Age', 'Nationality', 'Team', 'Contract', 'Positions', 'Height(ft)', 'Weight(lbs)', 'Foot', 'Wage(€)', 'Total Stats', 'Base Stats']

df_filtered = df.filter(items=relevant_columns)

df_filtered.head()
```

```python
# Cleaning the 'Height(ft)', 'Weight(lbs)', and 'Wage(€)' columns
df['Height(ft)'] = df['Height(ft)'].str.replace('\"', '', regex=False)
df['Weight(lbs)'] = df['Weight(lbs)'].str.replace('lbs', '', regex=False)
df['Wage(€)'] = df['Wage(€)'].str.replace('€', '', regex=False)
df['Wage(€)'] = df['Wage(€)'].str.replace('K', '000', regex=False)

# Renaming the 'Value' column to 'Value(€)'
df.rename(columns={'Value': 'Value(€)'}, inplace=True)
df['Value(€)'] = df['Value(€)'].str.replace('€', '', regex=False)

# Specifying the relevant columns to filter
relevant_columns = ['Overall Rating', 'Name', 'Age', 'Nationality', 'Team', 'Contract', 
                    'Positions', 'Height(ft)', 'Weight(lbs)', 'Foot', 'Value(€)', 
                    'Wage(€)', 'Total Stats', 'Base Stats']

# Filtering the DataFrame to show only the relevant columns
df_filtered = df.filter(items=relevant_columns)

# Display the filtered DataFrame
df_filtered.head()
```

```python
# Rows where the value contains a full stop (.)
value_with_full_stop = df[df['Value(€)'].str.contains('.', regex=False)]
print(value_with_full_stop[['Value(€)']])
```

```python


value_with_K = df[df['Value(€)'].str.contains('K', regex=False)]
print(value_with_K[['Value(€)']])
```

```python
# Tranforming the Value column
def convert_value_to_currency(Value):
    if 'M' in Value:
        return float(Value.replace('M', '').replace('.', '')) * 1_000_000
    elif 'K' in Value:
        return float(Value.replace('K', '').replace('.', '')) * 1_000
    else:
        return float(Value)

# Apply the function to the 'Value(€)' column
df['Value(€)'] = df['Value(€)'].apply(convert_value_to_currency)
```

```python
# Tranforming the Release Clause column
df['Release Clause(€)'] = df['Release Clause(€)'].str.replace('€', '', regex=False)

def convert_release_clause_to_currency(Value):
    if 'M' in Value:
        return float(Value.replace('M', '').replace('.', '')) * 1_000_000
    elif 'K' in Value:
        return float(Value.replace('K', '').replace('.', '')) * 1_000
    else:
        return float(Value)

# Apply the function to the 'Value(€)' column
df['Release Clause(€)'] = df['Release Clause(€)'].apply(convert_release_clause_to_currency)
```

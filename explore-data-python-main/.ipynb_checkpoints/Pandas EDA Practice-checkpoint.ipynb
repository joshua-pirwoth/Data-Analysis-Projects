# Answering questions

```python
# The top 10 most committed crimes
top_10_crimes = crime.OFFENSE_CODE_GROUP.value_counts()[:10]
top_10_crimes
```

```python
# Percentage of the top 10 most committed crimes
display(top_10_crimes / crime.shape[0])
```

```python
# Bar plot of the above slice
(top_10_crimes / crime.shape[0]).plot.barh(title = 'Top 10 Offense Groups (as % of All Crimes)', ylabel = '')
```

```python
# The least 10 most committed crimes
least_committed_crimes = crime.OFFENSE_CODE_GROUP.value_counts().sort_values(ascending = True)[:10]
least_committed_crimes
```

```python
# In which year were the most crimes created?
crime.groupby('YEAR').count()['INCIDENT_NUMBER'].plot(kind = 'bar', title = 'Number of crimes Committed Each Year')
```

```python
# Are there more crimes committed on specific days

# Ensuring that the days of the week are rendered in their logical order
crime['DAY_OF_WEEK'] = pd.Categorical(crime['DAY_OF_WEEK'], categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], ordered = True)

```

```python
crime
```

```python
crime.groupby('DAY_OF_WEEK', observed=False).count()
```

```python
# Crimes committed per specific day in a specific month
crime.groupby(['MONTH', 'DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
```

```python
day_and_month = crime.groupby(['MONTH', 'DAY_OF_WEEK'], observed = False).count()['INCIDENT_NUMBER'].unstack()
```

```python
day_and_month = day_and_month[['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']]
```

```python
sns.heatmap(day_and_month, cmap=sns.cubehelix_palette(as_cmap=True));
```

```python
day_and_month = crime.groupby(['MONTH', 'DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack('MONTH')
```

```python
# Map month numbers to their corresponding names
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
day_and_month.columns = [month_names[i-1] for i in day_and_month.columns]
```

```python
sns.heatmap(day_and_month, cmap=sns.cubehelix_palette(as_cmap=True));
plt.title("Crimes committed per specific day in a specific month")
plt.ylabel("")
plt.xlabel("")
```
